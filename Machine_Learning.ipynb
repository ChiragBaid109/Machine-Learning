Q1) Explain the differences between AI, ML, Deep Learning (DL), and Data
Science (DS).

ans-Artificial Intelligence (AI):
AI is the broadest concept, representing the overarching goal of creating machines that can perform tasks traditionally requiring human intelligence. This includes capabilities such as problem-solving, learning, decision-making, perception, and understanding language. AI encompasses various approaches, including rule-based systems, expert systems, and more modern data-driven methods like machine learning.
Machine Learning (ML):
ML is a subset of AI that focuses on enabling systems to learn from data without being explicitly programmed. ML algorithms identify patterns and make predictions or decisions based on the data they are trained on. This includes various techniques like supervised learning (e.g., classification, regression), unsupervised learning (e.g., clustering), and reinforcement learning.
Deep Learning (DL):
DL is a specialized subset of ML that utilizes artificial neural networks with multiple layers (deep neural networks) to model complex patterns in data. Inspired by the structure and function of the human brain, DL excels at tasks involving large, complex datasets, such as image recognition, natural language processing, and speech recognition, by automatically extracting features from raw data.
Data Science (DS):
DS is an interdisciplinary field that involves extracting knowledge and insights from structured and unstructured data. It combines principles from statistics, computer science, and domain expertise to analyze, interpret, and communicate data-driven findings. While data science often utilizes ML and DL techniques for analysis and prediction, it also encompasses other aspects like data collection, cleaning, visualization, statistical modeling, and experimental design, with a primary focus on deriving actionable insights for human understanding and decision-making.

Q2): What are the types of machine learning? Describe each with one
real-world example.

ans-The four main types of machine learning are Supervised Learning, Unsupervised Learning, Semi-Supervised Learning, and Reinforcement Learning. Supervised learning uses labeled data for prediction (e.g., spam filters), unsupervised learning identifies patterns in unlabeled data (e.g., customer segmentation), semi-supervised learning combines labeled and unlabeled data (e.g., medical image analysis), and reinforcement learning trains models through trial-and-error with rewards (e.g., a robot learning to walk).

Q3) : Define overfitting, underfitting, and the bias-variance tradeoff in machine learning

ans-Underfitting occurs when a model is too simple to capture the underlying patterns in the data, resulting in high bias and poor performance on both training and test data. Overfitting happens when a model is too complex and learns the training data, including noise, too well, leading to high variance and poor generalization to new data. The bias-variance tradeoff is the fundamental challenge of finding a model complexity that balances high bias (underfitting) with high variance (overfitting) to achieve optimal performance on unseen data

Q4) What are outliers in a dataset, and list three common techniques for
handling them

ans- Outliers in a dataset are data points that significantly deviate from other observations. They are values that are exceptionally far from the majority of the data points, either extremely high or extremely low. Outliers can arise from various sources, including measurement errors, data entry mistakes, natural variation, or rare events. Their presence can disproportionately influence statistical analyses and machine learning models, potentially leading to inaccurate conclusions.

```python
"""8 Write a Python script to:
● Generate 1000 samples from a normal distribution.
● Introduce 50 random missing values.
● Fill missing values with the column mean.
● Plot a histogram before and after imputation
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate 1000 samples from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=0, scale=1, size=1000)

# Step 2: Introduce 50 random missing values
missing_indices = np.random.choice(len(data), size=50, replace=False)
data_with_nan = data.copy()
data_with_nan[missing_indices] = np.nan

# Step 3: Fill missing values with the column mean
df = pd.DataFrame({'values': data_with_nan})
mean_value = df['values'].mean()
df_filled = df.fillna(mean_value)

# Step 4: Plot histograms before and after imputation
plt.figure(figsize=(12, 5))

# Before imputation
plt.subplot(1, 2, 1)
plt.hist(df['values'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title("Before Imputation (Missing Removed)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# After imputation
plt.subplot(1, 2, 2)
plt.hist(df_filled['values'], bins=30, color='lightgreen', edgecolor='black')
plt.title("After Imputation (Mean Filled)")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
